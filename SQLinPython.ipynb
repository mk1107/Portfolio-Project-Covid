```python
# importing csv to read csv data, panda for data manupulation and pandasql for running sql queries in python

import csv
import pandas as pd
from pandasql import sqldf
```

```python
#importing data in excel format

df1 = pd.read_excel('data/CovidDeaths.xlsx')
df2 = pd.read_excel('data/CovidVaccinations.xlsx')
```

```python
# CovidDeaths.xlsx

output = sqldf("SELECT * FROM df1")
output
```

```python
# Selecting data to be used 

output = sqldf("""SELECT location, DATE(date) AS date, total_cases, new_cases, total_deaths, population 
                    FROM df1
                    WHERE continent IS NOT null 
                    ORDER BY location, date""")
output
```

```python
# Total cases vs total deaths
# Liklyhood of dying due to covin in India

output = sqldf("""SELECT location, DATE(date) AS date, total_cases, total_deaths, (total_deaths/total_cases)*100 AS DeathPercentage 
                FROM df1
                WHERE location = 'India' AND continent IS NOT null 
                ORDER BY 1,2""")
output
```

```python
# Total cases vs population
# Precentage of population got covid in India

output = sqldf("""SELECT location, DATE(date) AS date, population, total_cases, (total_cases/population)*100 AS CasePercentage 
                FROM df1 
                WHERE location = 'India' AND continent IS NOT null 
                ORDER BY 1,2
                """)
output
```

```python
# Countries with highest infection rate compaired to population

output = sqldf("""SELECT location, population, MAX(total_cases) AS HighestInfectionCount, MAX(total_cases/population)*100 AS HighestInfectionPercentage 
                FROM df1
                WHERE continent IS NOT null 
                GROUP BY location, population 
                ORDER BY HighestInfectionPercentage desc
                """)
output
```

```python
# Countries wise highest death count

output = sqldf("""SELECT location, MAX(total_deaths) AS HighestDeath 
                FROM df1 
                WHERE continent IS NOT null 
                GROUP BY location 
                ORDER BY HighestDeath desc
                """)
output
```

```python
# Countries with highest death count per population

output = sqldf("""SELECT location, population, MAX(total_deaths) AS HighestDeathCount, MAX(total_deaths/population)*100 AS HighestdeathPercentage 
                FROM df1 
                WHERE continent IS NOT null 
                GROUP BY location, population 
                ORDER BY HighestDeathPercentage desc
                """)
output
```

```python
# Groupwise wise highest death count

output = sqldf("""SELECT location, MAX(total_deaths) AS HighestDeath 
                FROM df1
                WHERE continent IS null 
                GROUP BY location 
                ORDER BY HighestDeath desc
                """)
output
```

```python
# Continent wise highest death count

output = sqldf("""SELECT continent, MAX(total_deaths) AS HighestDeath 
                FROM df1 
                WHERE continent IS NOT null 
                GROUP BY continent 
                ORDER BY HighestDeath desc
                """)
output
```

```python
# Date wise Global Death precentage 

output = sqldf("""SELECT DATE(date) AS dat, SUM(new_cases) AS total_cases, SUM(new_deaths) AS total_deaths, (SUM(new_deaths)/SUM(new_cases))*100 AS DeathPercent 
                FROM df1 
                WHERE continent IS NOT null 
                GROUP BY date 
                ORDER BY 1,2
                """)
output
```

```python
# CovidVaccinations.xlsx

output = sqldf("SELECT * FROM df2")
output
```

```python
# Country wise vaccination Status

output = sqldf("""SELECT location, total_vaccinations, people_vaccinated AS PeopleWithFirstDose, people_fully_vaccinated AS PeopleWithSecondDose, total_boosters AS PeopleWithBoosterShot
                FROM df2
                WHERE continent IS NOT NULL 
                    AND total_vaccinations IN (
                        SELECT MAX(CAST(total_vaccinations AS decimal))
                        FROM df2
                        WHERE continent IS NOT NULL 
                        GROUP BY location
                    ) 
                ORDER BY CAST(total_vaccinations AS decimal) desc
                """)
output
```

```python
# Total Population vs Vaccinations
# Shows Percentage of Population that has recieved at least one Covid Vaccine

output = sqldf("""SELECT dea.continent, dea.location, DATE(dea.date) AS date, dea.population, vac.new_vaccinations, 
                        SUM(vac.new_vaccinations) OVER (PARTITION BY dea.Location ORDER BY dea.location, dea.Date) AS RollingPeopleVaccinated
                FROM df1 AS dea
                JOIN df2 AS vac
                    ON dea.location = vac.location
                    AND dea.date = vac.date
                WHERE dea.continent IS NOT NULL 
                ORDER BY 2,3
                """)
output
```
